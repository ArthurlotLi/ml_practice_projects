```python
# Harness wide constants
data_folder = "./data"
dataset_name = "car.data"
```

```python
# Basic imports

import numpy as np
import pandas as pd

pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
pd.set_option("display.width", None)
pd.set_option("display.max_colwidth", -1)
```

```python
# Read in data

data_names = ["buying","maint","doors","persons","lug_boot","safety","class"]
data_delimiter = ","

data = pd.read_csv(data_folder + "/" + dataset_name, names=data_names, delimiter=data_delimiter)

data
```

```python
# Data analysis

import matplotlib.pyplot as plt

data_colors = {"unacc":"red","acc":"blue","good":"green","vgood":"gold"}

def graph_attribute(data, dataset_name, attribute, solution):
    plt.title(dataset_name + " dataset - " + attribute)
    plt.scatter(data.index, data[attribute], color=data[solution].map(data_colors))
```

```python
for col in data:
    graph_attribute(data, "car_evaluation", col, "class")
    plt.show()
```

```python
# Dataset seperation

use_dataset_1 = True
use_dataset_2 = False

data_numpy = data.to_numpy()
X = None
Y = None

if use_dataset_1 is True:
    X = data_numpy[:,0:6]
    Y = data_numpy[:,6]
elif use_dataset_2 is True:
    X = data_numpy[:,0:6]
    Y = data_numpy[:,6]
    X = np.delete(X, 2, 1)
```

```python
# Encode all X variables

from sklearn.preprocessing import LabelEncoder

X_encoder = LabelEncoder()

# Note: make copies just in case changes are made to the code
# and we want to re-run the previous cells.
X_encoded = X.copy()

X_rows, X_cols = X_encoded.shape

for i in range(0, X_cols):
    X_encoded[:,i] = X_encoder.fit_transform(X_encoded[:,i])
    
X_encoded = X_encoded.astype("float")
```

```python
# One Hot encode Y variables

from keras.utils.np_utils import to_categorical

Y_encoder = LabelEncoder()
Y_encoded = Y_encoder.fit_transform(Y).astype("float")
Y_dummy = to_categorical(Y_encoded)
```

```python
# 
# Neural Network 
#

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

epochs = 500
batch_size = 32

Y_rows, Y_cols = Y_dummy.shape

def baseline_nn_model():
    model = Sequential()
    model.add(Dense(24, input_dim = X_cols, activation="relu"))
    model.add(Dense(8, activation="relu"))
    model.add(Dense(Y_cols, activation="softmax"))
    model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
    return model

from sklearn.pipeline import Pipeline
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import KFold, cross_val_score

estimators = []
estimators.append(("mlp",KerasClassifier(build_fn=baseline_nn_model, epochs=epochs,batch_size=batch_size)))
pipeline = Pipeline(estimators)
kfold = KFold(n_splits=10, shuffle=True)
```

```python
result_nn = cross_val_score(pipeline, X_encoded, Y_dummy, cv=kfold)
```

```python
print("Baseline Neural Network: %.2f (%.2f)" % (result_nn.mean()*100, result_nn.std()*100))
```

```python
# Make predictions with neural network

model_nn = pipeline.fit(X_encoded, Y_dummy)
```

```python
predictions_nn = model_nn.predict(X_encoded)
```

```python
# Display predictions made by the neural network model. 
predictions_nn = model_nn.predict(X_encoded)

# Note what we get back is an array of INDICES populated with the 
# one in the one-hot-encoded arrays. 

# Let's map our output to actual classes. Note that this isn't
# a very good method in case the mapping changes for Y when
# generating Y_dummy. 
#predictions_nn_map = {0:"acc",1:"good",2:"unacc",3:"vgood"}
predictions_nn_unencoded = Y_encoder.inverse_transform(predictions_nn)

data_with_nn_predictions = data.copy()

data_with_nn_predictions["class_predicted"] = predictions_nn_unencoded

data_with_nn_predictions
```

```python
#
# K-Nearest Neighbor
# 
# (Refernece code from "get-your-hands-dirty-with-scikit-learn-now")

from sklearn.neighbors import KNeighborsClassifier

n_neighbors=5
metric="euclidean"

def baseline_knn_model():
    model = KNeighborsClassifier(n_neighbors=n_neighbors, metric=metric)
    return model

model_knn = baseline_knn_model()
kfold = KFold(n_splits=10, shuffle=True)
result_knn = cross_val_score(model_knn, X_encoded, Y_dummy, cv=kfold)
```

```python
print("Baseline K-Nearest Neighbor: %.2f (%.2f)" % (result_knn.mean()*100, result_knn.std()*100))
```

```python
# Display predictions made by the knn model
model_knn = model_knn.fit(X_encoded, Y_dummy)

# We get dummy arrays back, so we can use np.argmax to 
# get the indices of the maximum values along a particular
# axis, in this case, the columns. 
predictions_knn = model_knn.predict(X_encoded)
predictions_knn = np.argmax(predictions_knn, axis=1, out=None)

# Note what we get back is an array of INDICES populated with the 
# one in the one-hot-encoded arrays. 

# Let's map our output to actual classes. Note that this isn't
# a very good method in case the mapping changes for Y when
# generating Y_dummy. 
#predictions_nn_map = {0:"acc",1:"good",2:"unacc",3:"vgood"}
predictions_knn_unencoded = Y_encoder.inverse_transform(predictions_knn)

data_with_knn_predictions = data.copy()

data_with_knn_predictions["class_predicted"] = predictions_knn_unencoded

data_with_knn_predictions
```
