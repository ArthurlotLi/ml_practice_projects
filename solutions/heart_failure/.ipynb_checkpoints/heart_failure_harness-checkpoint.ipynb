```python
data_folder = "./data"
dataset_name = "heart_failure_clinical_records_dataset.csv"
```

```python
import numpy as np
import pandas as pd

pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
pd.set_option("display.width", None)
pd.set_option("display.max_colwidth", -1)
```

```python
# Import data

data = pd.read_csv(data_folder + "/" + dataset_name)

data
```

```python
# Graph data

import matplotlib.pyplot as plt

data_colors = {1:"red",0:"blue"}

def graph_attribute(data, dataset_name, attribute, solution):
    plt.title(dataset_name + " dataset - " + attribute)
    plt.scatter(data.index, data[attribute], color=data[solution].map(data_colors))
    plt.show()
  
for col in data:
    graph_attribute(data, "heart_failure", col, "DEATH_EVENT")
```

```python
# Create X and Y

data_numpy = data.to_numpy()

X = data_numpy[:,0:12].astype("float")
Y = data_numpy[:,12].astype("float")
```

```python
# Neural Network

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

epochs = 200
batch_size = 32

X_rows, X_cols = X.shape

def baseline_nn_model():
    model = Sequential()
    model.add(Dense(12, input_dim=X_cols, activation="relu"))
    #model.add(Dense(6, input_dim=X_cols, activation="relu"))
    model.add(Dense(1, activation="sigmoid"))
    model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
    return model

from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import cross_val_score, cross_val_predict, KFold
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier

estimators = []
estimators.append(("standardize", StandardScaler()))
estimators.append(("clf",KerasClassifier(build_fn=baseline_nn_model, epochs=epochs,batch_size=batch_size)))
pipeline = Pipeline(estimators)
kfold = KFold(n_splits=10, shuffle=True)
```

```python
# Neural Network - Evaluation
results_nn = cross_val_score(pipeline, X, Y, cv=kfold, scoring='accuracy')
```

```python
print("Baseline Neural Network: %0.2f (%0.2f) Accuracy" % (results_nn.mean()*100,results_nn.std()*100))
```

```python
# Neural Network - Predictions

model_nn = pipeline.fit(X, Y)
```

```python
predictions_nn = model_nn.predict(X)

data_with_predictions_nn = data.copy()
data_with_predictions_nn["DEATH_EVENT_predicted"] = predictions_nn
data_with_predictions_nn
```

```python
# K Nearest Neighbor 

from sklearn.neighbors import KNeighborsClassifier

n_neighbors = 5
metric = "manhattan"

def baseline_knn_model():
    model = KNeighborsClassifier(n_neighbors=n_neighbors, metric=metric)
    return model

estimators_knn = []
estimators_knn.append(("standardize", StandardScaler()))
estimators_knn.append(("knn",baseline_knn_model()))
pipeline_knn = Pipeline(estimators_knn)
kfold_knn = KFold(n_splits=10, shuffle=True)

results_knn = cross_val_score(pipeline_knn, X, Y, cv=kfold_knn)

print("Baseline KNN Model: %.2f (%.2f) Accuracy" % (results_knn.mean()*100,results_knn.std()*100))
```

```python
# K Nearest Neighbor - Predictions

model_knn = pipeline_knn.fit(X, Y)

predictions_knn = model_knn.predict(X)

data_with_predictions_knn = data.copy()
data_with_predictions_knn["DEATH_EVENT_predicted"] = predictions_knn
data_with_predictions_knn
```

```python
# Gaussian Naive Bayes 

from sklearn.naive_bayes import GaussianNB

def baseline_gnb_model():
    model = GaussianNB()
    return model

estimators_gnb = []
estimators_gnb.append(("standardize", StandardScaler()))
estimators_gnb.append(("clf",baseline_gnb_model()))
pipeline_gnb = Pipeline(estimators_gnb)
kfold_gnb = KFold(n_splits=10, shuffle=True)

results_gnb = cross_val_score(pipeline_gnb, X, Y, cv=kfold_gnb)

print("Baseline GNB Model: %.2f (%.2f) Accuracy" % (results_gnb.mean()*100,results_gnb.std()*100))
```
