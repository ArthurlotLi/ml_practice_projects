```python
# Banknote Authentication

data_folder = "./data"
dataset_name = "data_banknote_authentication.txt"

import numpy as np
import pandas as pd
```

```python
# Import Data

data_names = ["variance","skewness","curtosis","entropy","class"]
data_delimiter=","

data = pd.read_csv(data_folder + "/" + dataset_name, names=data_names, delimiter=data_delimiter)

data
```

```python
# Graphing Attributes

import matplotlib.pyplot as plt

data_colors = {0:"green", 1:"red"}
dataset_name = "banknote_authentication"

def scatter_attribute(data, attribute,solution):
    plt.title(dataset_name + " dataset - " + attribute)
    plt.scatter(data.index, data[attribute], color = data[solution].map(data_colors))
    plt.show()

for col in data:
    scatter_attribute(data, col, "class")
```

```python
# Data Preprocessing (not much)
#
# Note: this dataset looks... far too clean. There's no documentation
# as to which class is which (genuine/forged) either. It simply looks
# like the images were not varied, or perhaps the samples were not 
# very different?
#
# I'd like to have seen the actual images themselves, or at least some
# samples. Otherwise this problem feels rather suspiciously clean. 

data_numpy = data.to_numpy()

data_rows, data_cols = data_numpy.shape

X = data_numpy[:,0:data_cols-1].astype("float")
Y = data_numpy[:,data_cols-1].astype("float")
```

```python
# Model Spot Check - Known Classification Algorithms 

nn_epochs = 71
nn_batch_size = 10
knn_n_neighbors = 5
knn_metric = "euclidean"
dt_criterion="entropy"
rf_criterion="gini"
rf_n_estimators=100
rf_max_depth=None
svm_C=0.1
svm_kernel="rbf"

X_rows, X_cols = X.shape

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB, MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn_lvq import GlvqModel

def baseline_nn():
    model = Sequential()
    model.add(Dense(4, input_dim=X_cols, activation="relu"))
    model.add(Dense(1))
    model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
    return model

def baseline_knn():
    return KNeighborsClassifier(n_neighbors=knn_n_neighbors, metric=knn_metric)

def baseline_gnb():
    return GaussianNB()

def baseline_mnb():
    return MultinomialNB()

def baseline_dt():
    return DecisionTreeClassifier(criterion=dt_criterion)

def baseline_rf():
    return RandomForestClassifier(criterion=rf_criterion, max_depth=rf_max_depth, n_estimators=rf_n_estimators)

def baseline_svm():
    return SVC(C=svm_C, kernel=svm_kernel)

def baseline_lvq():
    return GlvqModel()

from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import KFold, cross_val_score
from sklearn.pipeline import Pipeline
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier

kfold = KFold(n_splits=10, shuffle=True)

def model_pipeline_evaluate_print(model, scaler, model_name):
    model_result_print(model_evaluation(model_pipeline(model, scaler)),model_name)

def model_pipeline(model, scaler):
    estimators = []
    estimators.append(("standardize",scaler))
    estimators.append(("clf", model))
    return Pipeline(estimators)

def model_evaluation(pipeline):
    return cross_val_score(pipeline, X, Y, cv=kfold, scoring="accuracy")

def model_result_print(result, model_name):
    print("Baseline " + model_name + ": %.2f (%.2f)" % (result.mean()*100,result.std()*100))
```

```python
model_pipeline_evaluate_print(baseline_knn(), StandardScaler(), "KNN")
model_pipeline_evaluate_print(baseline_gnb(), MinMaxScaler(feature_range=(0,1)), "GNB")
model_pipeline_evaluate_print(baseline_mnb(), MinMaxScaler(feature_range=(0,1)), "MNB")
model_pipeline_evaluate_print(baseline_dt(), StandardScaler(), "DT")
model_pipeline_evaluate_print(baseline_rf(), StandardScaler(), "RF")
model_pipeline_evaluate_print(baseline_svm(), StandardScaler(), "SVM")
model_pipeline_evaluate_print(baseline_lvq(), StandardScaler(), "LVQ")
```

```python
result_nn = model_evaluation(model_pipeline(KerasClassifier(build_fn=baseline_nn, epochs=nn_epochs, batch_size=nn_batch_size),StandardScaler()))
```

```python
model_result_print(result_nn, "NN")
```

```python
# New Algorithm - Gradient Boosting

from sklearn.ensemble import GradientBoostingClassifier

gb_learning_rate = 0.1 # Default
gb_n_estimators = 100 # Same thing as RF
gb_subsample = 1.0 # Default 1.0. Fraction of samples to be used...
gb_criterion = "friedman_mse" # Default. Quality of the split. 

# Note! The fact we use subsampling means we're actually using
# Stochastic Gradient Boosting (same principle as RF's trees)

# Note 2: We're using MSE even though it's a classifier because
# Gradient Boosting uses Regression Trees as opposed to CART/DT
# trees so we can actually execute the gradient descent optimization.

def baseline_gb():
    return GradientBoostingClassifier(learning_rate = gb_learning_rate, n_estimators=gb_n_estimators, subsample=gb_subsample, criterion=gb_criterion)

model_pipeline_evaluate_print(baseline_gb(), StandardScaler(), "GB")
```
