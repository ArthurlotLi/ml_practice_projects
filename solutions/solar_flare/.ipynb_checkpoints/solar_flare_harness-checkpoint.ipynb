```python
data_folder = "./data"
dataset_names = ["flare.data1","flare.data2"]
#dataset_names = ["flare.data2"]

import numpy as np
import pandas as pd

pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
pd.set_option("display.width", None)
pd.set_option("display.max_colwidth", None)
```

```python
data_names = ["class","largest_spot","spot_distribution","activity","evolution","prev_24_hr","historically_complex","historically_complex_this_pass","area","area_largest_spot","c_num","m_num","x_num"]

dataset_files = []
for file in dataset_names:
    dataset_files.append(data_folder + "/" + file)

data = pd.concat((pd.read_csv(f, delim_whitespace=True, names=data_names) for f in dataset_files))

# Design choice: We want to predict EITHER X or M class flares, ignoring
# common class flares to alert people for further scrutinization of the
# sun in the coming hours. 
sol_col = []
# for fun
x_total = 0
m_total = 0
data_rows, data_cols = data.shape
for i in range(0, data_rows):
    row = data.iloc[i]
    row_x = row["x_num"]
    row_m = row["m_num"]
    if(row_x > 0): x_total = x_total + row_x
    if(row_m > 0): m_total = m_total + row_m
    sol_col.append(row_x + row_m)
    
data["mx_num"] = sol_col

data
```

```python
m_total, x_total
```

```python
import matplotlib.pyplot as plt

# Since we know the distribution of data, and there's so few
# numbers, we don't need to go through a color bar with cmap
# and boundaries. 
data_colors = {0:"blue",1:"red",2:"yellow",3:"green",4:"purple",5:"brown",6:"cyan",8:"black"}

def graph_attribute(data, dataset_name, attribute, solution):
    plt.title(dataset_name + " dataset - " + attribute)
    plt.scatter(data.index, data[attribute], color=data[solution].map(data_colors))
    plt.show()

for col in data:
    graph_attribute(data, "solar_flare", col, "mx_num")
```

```python
# Data preprocessing

data_processed = data.copy()

data_processed = data_processed.drop("x_num", axis=1)
data_processed = data_processed.drop("m_num", axis=1)
data_processed = data_processed.drop("c_num", axis=1)

data_numpy = data_processed.to_numpy()

data_numpy_rows, data_numpy_cols = data_numpy.shape

X = data_numpy[:, 0:data_numpy_cols-1]
Y = data_numpy[:, data_numpy_cols-1].astype("float")
```

```python
# One Hot Encoding for cols 0, 1, and 2.

from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

X_encoded = X.copy()

for i in range(0,3):
    X_encoded[:,i] = encoder.fit_transform(X_encoded[:,i])
    
X_encoded = X_encoded.astype("float")
```

```python
# Model Spot Check - known models

nn_epochs = 250
nn_batch_size = 10
knn_n_neighbors = 5
knn_metric = "euclidean"
dt_criterion = "mse"
rf_n_estimators = 100 
rf_max_depth = None
rf_criterion="mse"

X_encoded_rows, X_encoded_cols = X_encoded.shape

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.neighbors import KNeighborsRegressor
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression

def baseline_nn():
    model = Sequential()
    model.add(Dense(12, input_dim=X_encoded_cols, activation="relu"))
    model.add(Dense(1))
    model.compile(loss="mean_squared_error", optimizer="adam", metrics=["mse"])
    return model

def baseline_knn():
    model = KNeighborsRegressor(n_neighbors=knn_n_neighbors, metric=knn_metric)
    return model

def baseline_gnb():
    model = GaussianNB()
    return model

def baseline_dt():
    model = DecisionTreeRegressor(criterion=dt_criterion)
    return model

def baseline_rf():
    model = RandomForestRegressor(n_estimators = rf_n_estimators, max_depth = rf_max_depth, criterion=rf_criterion)
    return model

def baseline_lr():
    model = LinearRegression()
    return model

# Create pipelines

from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from tensorflow.keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import KFold, cross_val_score

kfold = KFold(n_splits = 10, shuffle=True)

estimators_nn = []
estimators_knn = []
estimators_gnb = []
estimators_dt = []
estimators_rf = []
estimators_lr = []

estimators_nn.append(("standardize",StandardScaler()))
estimators_nn.append(("mlp",KerasRegressor(build_fn=baseline_nn, epochs=nn_epochs, batch_size=nn_batch_size)))
pipeline_nn = Pipeline(estimators_nn)

estimators_knn.append(("standardize",StandardScaler()))
estimators_knn.append(("mlp",baseline_knn()))
pipeline_knn = Pipeline(estimators_knn)

estimators_gnb.append(("standardize",StandardScaler()))
estimators_gnb.append(("mlp",baseline_gnb()))
pipeline_gnb = Pipeline(estimators_gnb)

estimators_dt.append(("standardize",StandardScaler()))
estimators_dt.append(("mlp",baseline_dt()))
pipeline_dt = Pipeline(estimators_dt)

estimators_rf.append(("standardize",StandardScaler()))
estimators_rf.append(("mlp",baseline_rf()))
pipeline_rf = Pipeline(estimators_rf)

estimators_lr.append(("standardize",StandardScaler()))
estimators_lr.append(("mlp",baseline_lr()))
pipeline_lr = Pipeline(estimators_lr)
```

```python
result_nn = cross_val_score(pipeline_nn, X_encoded, Y, cv=kfold, scoring="neg_mean_squared_error")
```

```python
print("Baseline Neural Network: %.4f (%.2f)" % (result_nn.mean(),result_nn.std()))
```

```python
result_knn = cross_val_score(pipeline_knn, X_encoded, Y, cv=kfold, scoring="neg_mean_squared_error")
print("Baseline KNN: %.4f (%.2f)" % (result_knn.mean(),result_knn.std()))
```

```python
result_gnb = cross_val_score(pipeline_gnb, X_encoded, Y, cv=kfold, scoring="neg_mean_squared_error")
print("Baseline GNB: %.4f (%.2f)" % (result_gnb.mean(),result_gnb.std()))
```

```python
result_dt = cross_val_score(pipeline_dt, X_encoded, Y, cv=kfold, scoring="neg_mean_squared_error")
print("Baseline DT: %.4f (%.2f)" % (result_dt.mean(),result_dt.std()))
```

```python
result_rf = cross_val_score(pipeline_rf, X_encoded, Y, cv=kfold, scoring="neg_mean_squared_error")
print("Baseline RF: %.4f (%.2f)" % (result_rf.mean(),result_rf.std()))
```

```python
result_lr = cross_val_score(pipeline_lr, X_encoded, Y, cv=kfold, scoring="neg_mean_squared_error")
print("Baseline LR: %.4f (%.2f)" % (result_lr.mean(),result_lr.std()))
```

```python
# New Algorithm (kinda): Lasso Regression
# Least Absolute Shrinkage and Selection Operator Regularization

from sklearn.linear_model import Lasso

l_lr_alpha = 0.001 # Default is 1.0, aka lambda. Constant that multiples L1 term.

def baseline_l_lr():
    model = Lasso(alpha=l_lr_alpha)
    return model

estimators_l_lr = []
estimators_l_lr.append(("standardize",StandardScaler()))
estimators_l_lr.append(("mlp",baseline_l_lr()))
pipeline_l_lr = Pipeline(estimators_l_lr)

result_l_lr = cross_val_score(pipeline_l_lr, X_encoded, Y, cv=kfold, scoring="neg_mean_squared_error")
print("Baseline L_LR: %.4f (%.2f)" % (result_l_lr.mean(),result_l_lr.std()))
```
