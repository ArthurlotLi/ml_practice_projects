```python
# Thoracic Surgery Dataset

data_folder = "./data"
dataset_name = "ThoraricSurgery.data"

import pandas as pd
import numpy as np
```

```python
# Import data

data_names = ["DGN","PRE4","PRE5","PRE6","PRE7","PRE8","PRE9","PRE10","PRE11","PRE14","PRE17","PRE19","PRE25","PRE30","PRE32","AGE","class"]
data_delimiter=","

data = pd.read_csv(data_folder + "/" + dataset_name, delimiter=data_delimiter, names=data_names)

data
```

```python
# Graph attributes

import matplotlib.pyplot as plt

data_colors = {"F":"blue", "T": "red"}
data_alphas = {"F":0.1, "T":1.0 }

def scatter_attribute(data, dataset_name, attribute, solution):
    plt.title(dataset_name + " dataset - " + attribute)
    plt.scatter(data.index, data[attribute], color=data[solution].map(data_colors), alpha = data[solution].map(data_alphas))
    plt.show()

for col in data:
    scatter_attribute(data, "thoracic_surgery", col, "class")
```

```python
# Data Preprocessing:
# - Encode all cols that aren't int (i.e. all the T, F columns)

data_numpy = data.to_numpy()

data_rows, data_cols = data_numpy.shape

X = data_numpy[:,0:data_cols-1]
Y = data_numpy[:,data_cols-1]

from sklearn.preprocessing import LabelEncoder

X_rows, X_cols = X.shape
encoder = LabelEncoder()

for i in range(0,X_cols):
    encode_col = False
    try:
        test_col = X[:,i].astype("float")
    except:
        encode_col = True    
    if encode_col:
        X[:,i] = encoder.fit_transform(X[:,i])
        
Y = encoder.fit_transform(Y).astype("float")
X = X.astype("float")
```

```python
# Model Spot Check - Known Classification Algorithms 

nn_epochs = 150
nn_batch_size = 5
knn_n_neighbors = 5
knn_metric = "manhattan"
dt_criterion="entropy"
rf_criterion="gini"
rf_n_estimators=100
rf_max_depth=None
svm_C=1.5
svm_kernel="rbf"
cnn_epochs = 200
cnn_batch_size = 8

X_rows, X_cols = X.shape

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv1D
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB, MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

def baseline_nn():
    model = Sequential()
    model.add(Dense(8, input_dim=X_cols, activation="relu"))
    model.add(Dense(1))
    model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
    return model
def baseline_knn():
    return KNeighborsClassifier(n_neighbors=knn_n_neighbors, metric=knn_metric)
def baseline_gnb():
    return GaussianNB()
def baseline_mnb():
    return MultinomialNB()
def baseline_dt():
    return DecisionTreeClassifier(criterion=dt_criterion)
def baseline_rf():
    return RandomForestClassifier(criterion=rf_criterion,max_depth=rf_max_depth,n_estimators=rf_n_estimators)
def baseline_svm():
    return SVC(C=svm_C)
def baseline_cnn():
    model = Sequential()
    model.add(Conv1D(12, (1), input_shape=X_cols, activation="relu"))
    model.add(Dense(1))
    model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
    return model

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import KFold, cross_val_score
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier

kfold = KFold(n_splits=10, shuffle=True)

def model_pipeline(model):
    estimators=[]
    estimators.append(("standardize",StandardScaler()))
    estimators.append(("clf", model))
    return Pipeline(estimators)

def model_pipeline_minmax(model):
    estimators=[]
    estimators.append(("standardize",MinMaxScaler(feature_range=(0,1))))
    estimators.append(("clf", model))
    return Pipeline(estimators)

def model_evaluate(pipeline):
    return cross_val_score(pipeline, X, Y, cv=kfold, scoring="accuracy")

def model_print(result, model_name):
    print("Baseline " + model_name +": %.2f (%.2f)" % (result.mean()*100, result.std()*100))
    
def model_evaluate_print(model, model_name):
    model_print(model_evaluate(model_pipeline(model)), model_name)
    
```

```python
# Neural Network

result_nn = model_evaluate(model_pipeline(KerasClassifier(build_fn=baseline_nn, epochs=nn_epochs, batch_size=nn_batch_size)))
```

```python
model_print(result_nn, "NN")
```
