```python
# Drug Consumption

data_folder = "./data"
dataset_name = "drug_consumption.data"

import pandas as pd
import numpy as np

pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
pd.set_option("display.width", None)
pd.set_option("display.max_colwidth", None)
```

```python
# Import Data

data_names = ["id","age","gender","education","country","ethnicity","neuroticism","extraversion","open_experiences","agreeableness","conscientiousness","impulsiveness","sensation_seeing","alcohol","amphet","amyl","benzos","caffeine","cannabis","chocolate","cocaine","crack","ecstasy","heroin","ketamine","legal_highs","lsd","meth","mushrooms","nicotine","fake","volatile_substance"]
data_delimiter=","

data = pd.read_csv(data_folder + "/" + dataset_name, delimiter=data_delimiter, names=data_names)

data
```

```python
# Define solution column...
# 
# A few options I'm interested in for the solution. 
#
# 1. Have they done hard drugs y/n (Binary classification)
# 2. How long ago have they drunk alcohol (or any other, specific drug?)
# 3. What is the "hardest" drug they've taken, if any?
# 4. Have they done hard drugs in the lsat year (day, week, month, etc.)

option_selection = 4

data_modified = data.copy()
data_rows, data_cols = data_modified.shape
solution_column = []

if(option_selection == 1):
    print("SOLUTION OPTION 1 - HARD DRUGS OR NOT")
    
    drugs_counted = ["mushrooms","volatile_substance","ecstasy","lsd","crack","amphet","ketamine","benzos","amyl","cocaine","meth","heroin","fake"]
    drugs_unused = ["chocolate", "caffeine", "nicotine", "alcohol", "legal_highs","cannabis", "mushrooms", "volatile_substance"]
    
    for i in range(0,data_rows):
        row = data_modified.iloc[i]
        drug_taken = False
        for j in range(len(drugs_counted)-1,-1,-1):
            drug_to_check = drugs_counted[j]
            if(row[drug_to_check] != "CL0"):
                drug_taken = True
                break
        if drug_taken:
            solution_column.append("1")
        else:
            solution_column.append("0")
    
    # Drop the columns we won't be using anymore. 
    for col in drugs_counted:
        data_modified = data_modified.drop(col, axis=1)
    for col in drugs_unused:
        data_modified = data_modified.drop(col, axis=1)
        
    frequency_dict = {i:solution_column.count(i) for i in solution_column}
    print("\nFrequency:")
    print(frequency_dict)
    
elif(option_selection == 2):
    pass

elif(option_selection == 3):
    print("SOLUTION OPTION 3 - THE HARDEST DRUG")
    # The most a resulting class can have before if it is to be merged
    # with other drugs. Does not include "none"
    #
    # We have this so that we can have a good dataset that algorithms
    # can train on, at the expense of losing some information in the
    # final solution. 
    class_merge_cutoff = 100
    
    # Define what are the "hardest" drugs in sorted order.
    # We will then keep the "hardest" as the solution and ignore
    # all other possible drugs. 
    #
    # Disclaimer... I'm not an expert on drugs. Order roughly
    # obtained from DEA's drug-scheduling + reddit opinions. 
    # Chocolate, Caffeine, Nicotine, and Alcohol, Legal_Highs have 
    # been excluded from list, counting as "none".
    drugs_sorted = ["cannabis","mushrooms","volatile_substance","ecstasy","lsd","crack","amphet","ketamine","benzos","amyl","cocaine","meth","heroin","fake"]
    drugs_unused = ["chocolate", "caffeine", "nicotine", "alcohol", "legal_highs"]
    
    for i in range(0,data_rows):
        row = data_modified.iloc[i]
        hardest_drug = None
        for j in range(len(drugs_sorted)-1,-1,-1):
            drug_to_check = drugs_sorted[j]
            if(row[drug_to_check] != "CL0"):
                hardest_drug = drug_to_check
                break
        if hardest_drug is None:
            solution_column.append("none")
        elif hardest_drug == "fake":
            # We'll manage liars like this. 
            solution_column.append("none")
        else:
            solution_column.append(hardest_drug)
    
    # Drop the columns we won't be using anymore. 
    for col in drugs_sorted:
        data_modified = data_modified.drop(col, axis=1)
    for col in drugs_unused:
        data_modified = data_modified.drop(col, axis=1)
    
    frequency_dict = {i:solution_column.count(i) for i in solution_column}
    print("\nInitial Sorted Drugs Frequency:")
    print(frequency_dict)
    
    # Programatically group classes that have really small numbers of 
    # respondents. 
    grouped_drugs = []
    grouped_drugs_names = None
    for key in frequency_dict:
        if frequency_dict[key] < class_merge_cutoff and key != "none":
            grouped_drugs.append(key)
            drugs_sorted.remove(key)
            if(grouped_drugs_names is None):
                grouped_drugs_names = key
            else:
                grouped_drugs_names = grouped_drugs_names + "+" + key
    print("\nGrouped class: " + grouped_drugs_names)
    
    # Regenerate the grouped column with these grouped drugs.
    for i in range(0, len(solution_column)):
        if(solution_column[i] in grouped_drugs):
            solution_column[i] = grouped_drugs_names
            
    frequency_dict = {i:solution_column.count(i) for i in solution_column}
    print("\nFinal Sorted Drugs Frequency:")
    print(frequency_dict)
    print("\n[Total Participants: " + str(len(solution_column)) + "]")
    
elif(option_selection == 4):
    print("SOLUTION OPTION 4 - HARD DRUGS LAST DAY OR WEEK")
    
    drugs_counted = ["ecstasy","lsd","crack","amphet","ketamine","benzos","amyl","cocaine","meth","heroin","fake"]
    drugs_unused = ["chocolate", "caffeine", "nicotine", "alcohol", "legal_highs","cannabis", "mushrooms", "volatile_substance"]
    
    for i in range(0,data_rows):
        row = data_modified.iloc[i]
        drug_taken = False
        for j in range(len(drugs_counted)-1,-1,-1):
            drug_to_check = drugs_counted[j]
            if(row[drug_to_check] == "CL5" or row[drug_to_check] == "CL6"):
                drug_taken = True
                break
        if drug_taken:
            solution_column.append("1")
        else:
            solution_column.append("0")
    
    # Drop the columns we won't be using anymore. 
    for col in drugs_counted:
        data_modified = data_modified.drop(col, axis=1)
    for col in drugs_unused:
        data_modified = data_modified.drop(col, axis=1)
        
    frequency_dict = {i:solution_column.count(i) for i in solution_column}
    print("\nFrequency:")
    print(frequency_dict)

# Remove certain columns we will never use. 
data_modified = data_modified.drop("id", axis=1)

data_modified["class"] = solution_column
```

```python
# Data Analysis

import matplotlib.pyplot as plt
import random

# Automatically generate colors for each specific class, since the
# solution column may have any number of potential classes. 
data_classes = {i:solution_column.count(i) for i in solution_column}
data_classes_names = list(data_classes.keys())

# Use random colors. Note that there is the possiblity of 
# conflicts (or really poor color choices) so be sure to roll the
# dice enough times to generate nice enough graphs.  
colors = ["#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)])
             for i in range(len(data_classes_names))]
data_colors = {}
for i in range(0, len(data_classes_names)):
    data_colors[data_classes_names[i]] = colors[i]
    

def graph_attribute(data, attribute, solution):
    plt.title("drug_consumption - " + attribute)
    plt.scatter(data.index, data[attribute], color = data[solution].map(data_colors))
    plt.show()

for col in data_modified:
    graph_attribute(data_modified, col, "class")
```

```python
# The country graph was surprising as heck, so here's a histogram.

hist_legend = list(data_classes.keys())

def hist_attribute(data, attribute, solution):
    fig, ax = plt.subplots()
    ax.title.set_text("drug_consumption - " + attribute)
    graph_data = data.groupby(data[solution])[attribute]
    graph_data.hist(ax = ax, alpha=0.2,histtype='stepfilled', edgecolor='none')
    #ax.legend((hist_legend), loc='upper right')
    fig.set_size_inches(15, 4)

hist_cols = ["class","age", "gender", "education", "country", "ethnicity"]
    
for col in data_modified:
    if col in hist_cols:
        hist_attribute(data_modified, col, "class")
```

```python
# Data Preprocessing

data_numpy = data_modified.to_numpy()

data_rows, data_cols = data_numpy.shape

X = data_numpy[:,0:data_cols-1].astype("float")
Y = data_numpy[:,data_cols-1]

# One hot encoding for Y

from sklearn.preprocessing import LabelEncoder
from keras.utils.np_utils import to_categorical

encoder = LabelEncoder()
Y_encoded = encoder.fit_transform(Y).astype("float")
Y_dummy = to_categorical(Y_encoded).astype("float")
```

```python
X.shape
```

```python
# Model Spot Check - Known Algorithms 

nn_epochs = 200
nn_batch_size = 32
knn_n_neighbors = 4
knn_metric = "euclidean"
dt_criterion="entropy"
rf_criterion="entropy"
rf_max_depth=None
rf_n_estimators=400
svm_C=1.0
svm_kernel="rbf"
gb_criterion="friedman_mse"
gb_n_estimators=150
gb_subsample=1.0
gb_learning_rate=0.1
lda_solver="eigen"
lda_shrinkage="auto"

X_rows, X_cols = X.shape
Y_dummy_rows, Y_dummy_cols = Y_dummy.shape

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB, MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn_lvq import GlvqModel
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

def baseline_nn():
    model = Sequential()
    model.add(Dense(32, input_dim=X_cols, activation="relu"))
    model.add(Dense(16, activation="relu"))
    model.add(Dense(Y_dummy_cols, activation="softmax"))
    model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
    return model

def baseline_knn():
    return KNeighborsClassifier(n_neighbors=knn_n_neighbors, metric=knn_metric)

def baseline_gnb():
    return GaussianNB()

def baseline_mnb():
    return MultinomialNB()

def baseline_dt():
    return DecisionTreeClassifier(criterion=dt_criterion)

def baseline_rf():
    return RandomForestClassifier(criterion=rf_criterion, max_depth=rf_max_depth, n_estimators=rf_n_estimators)

def baseline_svm():
    return SVC(C=svm_C, kernel = svm_kernel)

def baseline_lvq():
    return GlvqModel()

def baseline_gb():
    return GradientBoostingClassifier(criterion=gb_criterion, n_estimators=gb_n_estimators, learning_rate=gb_learning_rate, subsample=gb_subsample)

def baseline_lda():
    return LinearDiscriminantAnalysis(solver=lda_solver, shrinkage=lda_shrinkage)

from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import cross_val_score, KFold
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier

kfold = KFold(n_splits=10, shuffle=True)

def model_pipeline(model, standardizer=StandardScaler()):
    estimators=[("standardize",standardizer)]
    estimators.append(("clf", model))
    return Pipeline(estimators)

def model_evaluate(pipeline, X_array, Y_array):
    return cross_val_score(pipeline, X_array, Y_array, cv=kfold, scoring="accuracy")

def model_print(result, model_name):
    print("Baseline " + model_name + ": %.2f (%.2f)" % (result.mean()*100, result.std()*100))
    
def model_full(model, model_name, X_array, Y_array, standardizer=StandardScaler(), return_result=False):
    pipeline = model_pipeline(model, standardizer)
    result = model_evaluate(pipeline, X_array, Y_array)
    model_print(result, model_name)
    if return_result:
        return result
```

```python
model_full(baseline_svm(), "svm", X, Y_encoded) 
```

```python
model_full(baseline_knn(), "knn", X, Y_encoded) 
model_full(baseline_gnb(), "gnb", X, Y_encoded, standardizer=MinMaxScaler(feature_range=(0,1))) 
model_full(baseline_mnb(), "mnb", X, Y_encoded, standardizer=MinMaxScaler(feature_range=(0,1))) 
model_full(baseline_dt(), "dt", X, Y_encoded) 
model_full(baseline_rf(), "rf", X, Y_encoded) 
model_full(baseline_svm(), "svm", X, Y_encoded) 
model_full(baseline_lvq(), "lvq", X, Y_encoded) 
model_full(baseline_gb(), "gb", X, Y_encoded) 
model_full(baseline_lda(), "lda", X, Y_encoded)
```

```python
#result_nn = model_full(KerasClassifier(build_fn=baseline_nn, epochs=nn_epochs, batch_size=nn_batch_size), "nn", X, Y_encoded, return_result=True, scoring = None)

estimators = []
estimators.append(("standardize",StandardScaler()))
estimators.append(("clf", KerasClassifier(build_fn=baseline_nn, epochs=nn_epochs, batch_size=nn_batch_size)))
pipeline_nn = Pipeline(estimators)

result_nn = cross_val_score(pipeline_nn, X, Y_dummy, cv=kfold)#, scoring="accuracy")
```

```python
model_print(result_nn, "nn")
```
