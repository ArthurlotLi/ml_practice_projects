```python
# Accent Recognition
#
# A sample dataset with limited applicability - a nice
# proof of concept exploration. 

data_folder = "./data"
dataset_name = "accent-mfcc-data-1.csv"

import numpy as np
import pandas as pd
```

```python
# Import Data

data = pd.read_csv(data_folder + "/" + dataset_name)

data
```

```python
solution_column = list(data["language"])

frequency_dict = {i:solution_column.count(i) for i in solution_column}
print(frequency_dict)
```

```python
# Data Analysis

import matplotlib.pyplot as plt

data_colors = {"ES":"red","FR":"blue","GE":"green","IT":"yellow","UK":"purple","US":"black"}
dataset_name = "accent_recognition"

def scatter_attribute(data, attribute, solution):
    plt.title(dataset_name + " dataset - " + attribute)
    plt.scatter(data.index, data[attribute], color=data[solution].map(data_colors))
    plt.show()
    
for col in data:
    scatter_attribute(data, col, "language")
```

```python
# Data Preprocessing

data_numpy = data.to_numpy()

data_rows, data_cols = data_numpy.shape

X = data_numpy[:,1:data_cols].astype("float")
Y = data_numpy[:,0]

# Encode Y

from sklearn.preprocessing import LabelEncoder
from keras.utils.np_utils import to_categorical

encoder = LabelEncoder()
Y_encoded = encoder.fit_transform(Y).astype("float")
Y_dummy = to_categorical(Y_encoded)
```

```python
# Model Spot Check - known algorithms

nn_epochs = 200
nn_batch_size = 5
knn_n_neighbors = 5
knn_metric = "euclidean"
dt_criterion = "entropy"
rf_criterion = "gini"
rf_n_estimators = 1000
rf_max_depth = None
svm_C = 1.0
svm_kernel = "rbf"
gb_criterion = "friedman_mse"
gb_n_estimators = 100
gb_learning_rate = 0.1
gb_subsample = 1.0

X_rows, X_cols = X.shape
Y_dummy_rows, Y_dummy_cols = Y_dummy.shape

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB, MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn_lvq import GlvqModel
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.linear_model import LogisticRegression

def baseline_nn():
    model = Sequential()
    model.add(Dense(12, input_dim = X_cols, activation="relu"))
    model.add(Dense(Y_dummy_cols, activation="softmax"))
    model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
    return model
    
def baseline_knn():
    return KNeighborsClassifier(n_neighbors = knn_n_neighbors, metric = knn_metric)

def baseline_gnb():
    return GaussianNB()

def baseline_mnb():
    return MultinomialNB()

def baseline_dt():
    return DecisionTreeClassifier(criterion=dt_criterion)

def baseline_rf():
    return RandomForestClassifier(criterion=rf_criterion, max_depth = rf_max_depth, n_estimators=rf_n_estimators)

def baseline_svm():
    return SVC(C=svm_C, kernel = svm_kernel)

def baseline_lvq():
    return GlvqModel()

def baseline_gb():
    return GradientBoostingClassifier(criterion=gb_criterion, n_estimators=gb_n_estimators, learning_rate=gb_learning_rate, subsample=gb_subsample)

# Whoops, we can't use logistic regression for a multi classification problem.
# ... Well, we could technically with one-or-all, but that would be too much effort. 
def baseline_log():
    return LogisticRegression()

def baseline_lda():
    return LinearDiscriminantAnalysis()

from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import KFold, cross_val_score
from sklearn.pipeline import Pipeline
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier

kfold = KFold(n_splits = 10, shuffle = True)

def model_pipeline(model, scaler=StandardScaler()):
    estimators=[("standardize",scaler), ("clf", model)]
    return Pipeline(estimators)

def model_evaluation(pipeline, X_array, Y_array, scoring="accuracy"):
    return cross_val_score(pipeline, X_array, Y_array, scoring=scoring)

def model_print(results, model_name):
    print("Baseline " + model_name + ": %.2f (%.2f)" % (results.mean()*100, results.std()*100))

def model_full(model, model_name, X_array, Y_array, scoring="accuracy", scaler=StandardScaler(), return_results=False):
    pipeline = model_pipeline(model, scaler)
    results = model_evaluation(pipeline, X_array, Y_array, scoring)
    model_print(results, model_name)
    if return_results:
        return results
```

```python
model_full(baseline_knn(), "knn", X, Y_encoded)
model_full(baseline_gnb(), "gnb", X, Y_encoded, scaler=MinMaxScaler(feature_range=(0,1)))
model_full(baseline_mnb(), "mnb", X, Y_encoded, scaler=MinMaxScaler(feature_range=(0,1)))
model_full(baseline_dt(), "dt", X, Y_encoded)
model_full(baseline_rf(), "rf", X, Y_encoded)
model_full(baseline_svm(), "svm", X, Y_encoded)
model_full(baseline_lvq(), "lvq", X, Y_encoded)
model_full(baseline_gb(), "gb", X, Y_encoded)
model_full(baseline_lda(), "lda", X, Y_encoded)
```

```python
results_nn = model_full(KerasClassifier(build_fn=baseline_nn, epochs=nn_epochs, batch_size=nn_batch_size), "nn", X, Y_dummy, return_results=True, scoring=None)
```

```python
model_print(results_nn, "nn")
```

```python
# New Concept - Singular Value Decomposition
# for dimensiality reduction (not too useful in
# this case, but good practice.)

from sklearn.decomposition import TruncatedSVD

# Fit the data using SVD. 
svd = TruncatedSVD()
X_transformed = svd.fit_transform(X)

X_transformed
```

```python
# Try running everything again using the decomposed data.

model_full(baseline_knn(), "knn", X_transformed, Y_encoded)
model_full(baseline_gnb(), "gnb", X_transformed, Y_encoded, scaler=MinMaxScaler(feature_range=(0,1)))
model_full(baseline_mnb(), "mnb", X_transformed, Y_encoded, scaler=MinMaxScaler(feature_range=(0,1)))
model_full(baseline_dt(), "dt", X_transformed, Y_encoded)
model_full(baseline_rf(), "rf", X_transformed, Y_encoded)
model_full(baseline_svm(), "svm", X_transformed, Y_encoded)
model_full(baseline_lvq(), "lvq", X_transformed, Y_encoded)
model_full(baseline_gb(), "gb", X_transformed, Y_encoded)
model_full(baseline_lda(), "lda", X_transformed, Y_encoded)
```

```python
X_transformed_rows, X_transformed_cols = X_transformed.shape

def baseline_nn_svd():
    model = Sequential()
    model.add(Dense(12, input_dim = X_transformed_cols, activation="relu"))
    model.add(Dense(Y_dummy_cols, activation="softmax"))
    model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
    return model

results_nn = model_full(KerasClassifier(build_fn=baseline_nn_svd, epochs=nn_epochs, batch_size=nn_batch_size), "nn", X_transformed, Y_dummy, return_results=True, scoring=None)
```

```python
model_print(results_nn, "nn")
```

```python
# New Concept - Principal Component Analysis
# SVD uses various numerical methods to derive
# the Singular Value Decomposition of the data. 
# PCA uses matrix manipulation to generate 
# eigenvectors and eigenvalues and selects the
# k eigenvectors with the largest eigenvalues.
#
# PCA is better for Dense datasets.
# SVD is better for Sparse datasets. 

from sklearn.decomposition import PCA

pca = PCA()
X_transformed_pca = pca.fit_transform(X)

# Result is out dataset back again. Turns out all of our
# collumns are good. 
X_transformed_pca.shape

X_transformed_pca
```

```python
# Regardless, let's try running everything again using the decomposed data.

model_full(baseline_knn(), "knn", X_transformed_pca, Y_encoded)
model_full(baseline_gnb(), "gnb", X_transformed_pca, Y_encoded, scaler=MinMaxScaler(feature_range=(0,1)))
model_full(baseline_mnb(), "mnb", X_transformed_pca, Y_encoded, scaler=MinMaxScaler(feature_range=(0,1)))
model_full(baseline_dt(), "dt", X_transformed_pca, Y_encoded)
model_full(baseline_rf(), "rf", X_transformed_pca, Y_encoded)
model_full(baseline_svm(), "svm", X_transformed_pca, Y_encoded)
model_full(baseline_lvq(), "lvq", X_transformed_pca, Y_encoded)
model_full(baseline_gb(), "gb", X_transformed_pca, Y_encoded)
model_full(baseline_lda(), "lda", X_transformed_pca, Y_encoded)
```

```python
X_transformed_pca_rows, X_transformed_pca_cols = X_transformed_pca.shape

def baseline_nn_pca():
    model = Sequential()
    model.add(Dense(12, input_dim = X_transformed_pca_cols, activation="relu"))
    model.add(Dense(Y_dummy_cols, activation="softmax"))
    model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
    return model

results_nn = model_full(KerasClassifier(build_fn=baseline_nn_pca, epochs=nn_epochs, batch_size=nn_batch_size), "nn", X_transformed_pca, Y_dummy, return_results=True, scoring=None)
```

```python
model_print(results_nn, "nn")
```
