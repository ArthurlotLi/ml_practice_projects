```python
data_folder = "./data"
dataset_names = ["xaa.dat","xab.dat","xac.dat","xad.dat","xae.dat","xaf.dat","xag.dat","xah.dat","xai.dat"]

import numpy as np
import pandas as pd

pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
pd.set_option("display.width", None)
pd.set_option("display.max_colwidth", -1)
```

```python
data_names = ["compactness","circularity","distance_circularity","radius_ratio","praxis_aspect_ratio","max_length_aspect_ratio","scatter_ratio","elongatedness","praxis_rectangulariy","maxlength_rectangularity","scaled_variance_major","scaled_variance_minor","scaled_radius_gyration","skewed_major","skewed_minor","kurtosis_minor","kurtosis_major","hollows","class"]

dataset_files = []
for file in dataset_names:
    dataset_files.append(data_folder + "/" + file)

data = pd.concat((pd.read_csv(f, names = data_names, delim_whitespace=True) for f in dataset_files))

data
```

```python
# Graphing attributes

import matplotlib.pyplot as plt

data_colors = {"van":"blue", "saab": "red", "opel": "purple", "bus": "black"}

def graph_attribute(data, dataset_name, attribute, solution):
    plt.title(dataset_name + " dataset - " + attribute)
    plt.scatter(data.index, data[attribute], color=data[solution].map(data_colors))
    plt.show()
    
for col in data:
    graph_attribute(data, "vehicle_silhouettes", col, "class")
```

```python
# Data seperation 

data_numpy = data.to_numpy()

X = data_numpy[:,0:18].astype("float")
Y = data_numpy[:,18]
```

```python
# One Hot Encoding for Y

from sklearn.preprocessing import LabelEncoder
from keras.utils.np_utils import to_categorical

encoder = LabelEncoder()
Y_encoded = encoder.fit_transform(Y)
Y_dummy = to_categorical(Y_encoded)
```

```python
# Model Spot Check - NN, KNN, GNB, DT

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier

nn_epochs = 200
nn_batch_size = 18
knn_metric = "manhattan"
knn_n_neighbors = 5
dt_criterion = "gini"

X_rows, X_cols = X.shape

def baseline_nn_model():
    model = Sequential()
    model.add(Dense(16, input_dim=X_cols, activation="relu"))
    model.add(Dense(4, activation="softmax"))
    model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
    return model

def baseline_knn_model():
    model = KNeighborsClassifier(n_neighbors=knn_n_neighbors, metric=knn_metric)
    return model

def baseline_gnb_model():
    model = GaussianNB()
    return model

def baseline_dt_model():
    model = DecisionTreeClassifier(criterion=dt_criterion)
    return model

from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import KFold, cross_val_score
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier

estimators_nn = []
estimators_nn.append(("standardize",StandardScaler()))
estimators_nn.append(("clf",KerasClassifier(build_fn=baseline_nn_model, epochs=nn_epochs, batch_size=nn_batch_size)))
pipeline_nn = Pipeline(estimators_nn)
kfold_nn = KFold(n_splits=10, shuffle=True)

estimators_knn = []
estimators_knn.append(("standardize",StandardScaler()))
estimators_knn.append(("clf",baseline_knn_model()))
pipeline_knn = Pipeline(estimators_knn)
kfold_knn = KFold(n_splits=10, shuffle=True)

estimators_gnb = []
estimators_gnb.append(("standardize",StandardScaler()))
estimators_gnb.append(("clf",baseline_gnb_model()))
pipeline_gnb = Pipeline(estimators_gnb)
kfold_gnb = KFold(n_splits=10, shuffle=True)

estimators_dt = []
estimators_dt.append(("standardize",StandardScaler()))
estimators_dt.append(("clf",baseline_dt_model()))
pipeline_dt = Pipeline(estimators_dt)
kfold_dt = KFold(n_splits=10, shuffle=True)
```

```python
results_nn = cross_val_score(pipeline_nn, X, Y_dummy, cv=kfold_nn)
```

```python
print("Baseline Neural Network: %0.2f (%0.2f)" % (results_nn.mean()*100,results_nn.std()*100))
```

```python
results_knn = cross_val_score(pipeline_knn, X, Y_encoded, cv=kfold_knn)
```

```python
print("Baseline KNN: %0.2f (%0.2f)" % (results_knn.mean()*100,results_knn.std()*100))
```

```python
results_gnb = cross_val_score(pipeline_gnb, X, Y_encoded, cv=kfold_gnb)
```

```python
print("Baseline Gaussian Naive Bayes: %0.2f (%0.2f)" % (results_gnb.mean()*100,results_gnb.std()*100))
```

```python
results_dt = cross_val_score(pipeline_dt, X, Y_encoded, cv=kfold_dt)
```

```python
print("Baseline Decision Tree: %0.2f (%0.2f)" % (results_dt.mean()*100,results_dt.std()*100))
```

```python
# New Algorithm - Random Forest

from sklearn.ensemble import RandomForestClassifier

rf_n_estimators = 100 # Number of trees in forest, default is 100.
rf_max_depth = None # Maximum depth of tree, default is none.
rf_criterion = "gini" # Default is gini

def baseline_rf_model():
    model = RandomForestClassifier(criterion = rf_criterion, n_estimators = rf_n_estimators, max_depth=rf_max_depth)
    return model

estimators_rf = []
estimators_rf.append(("standardize",StandardScaler()))
estimators_rf.append(("clf",baseline_rf_model()))
pipeline_rf = Pipeline(estimators_rf)
kfold_rf = KFold(n_splits=10, shuffle=True)

results_rf = cross_val_score(pipeline_rf, X, Y_encoded, cv=kfold_rf)
```

```python
print("Baseline Decision Tree: %0.2f (%0.2f)" % (results_rf.mean()*100,results_rf.std()*100))
```
