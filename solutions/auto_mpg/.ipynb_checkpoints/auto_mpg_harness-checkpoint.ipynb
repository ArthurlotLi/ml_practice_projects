```python
# Harness wide constants

data_folder = "./data"
dataset_name = "auto-mpg.data"

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', -1)
```

```python
# Basic imports

import numpy as np
import pandas as pd
```

```python
# Read in the data

data_names = ["mpg","cylinders","displacement","horsepower","weight","acceleration","model_year","origin","car_name"]

data = pd.read_csv(data_folder + "/" + dataset_name, names=data_names,delim_whitespace=True)

# For fun, let's add another column with the first word of the car_name. 
# Becuase certain brands suck at mpg. 

car_make = []
for name in data["car_name"]:
    car_make.append(name.split(" ")[0])
    
data["car_make"] = car_make

data
```

```python
data["mpg"].min(), data["mpg"].max()
```

```python
import matplotlib.pyplot as plt
from matplotlib.colors import BoundaryNorm, ListedColormap

boundaries = [9, 15, 20, 25, 30, 35, 40, 47]
colors = ["red","blue","yellow","cyan","purple","green","black"]
cmap = ListedColormap(colors)
norm = BoundaryNorm(boundaries, len(colors))

def graph_attribute(data, dataset_name, variable, solution):
    fig, ax = plt.subplots()
    ax.title.set_text(dataset_name + " dataset: " + variable)
    sc = ax.scatter(data.index, data[variable], c=data[solution], cmap=cmap, norm=norm)
    fig.colorbar(sc, spacing="proportional")
```

```python
for col in data:
    if col != "car_name":
        graph_attribute(data, "auto_mpg", col, "mpg")
```

```python
# Data preprocessing

from sklearn.preprocessing import LabelEncoder

use_dataset1 = True
use_dataset2 = False
use_dataset3 = False

data_numpy = data.to_numpy()
rows_to_delete = []

X = None
Y = None

if use_dataset1 or use_dataset3:
    # Remove the ? rows first because otherwise we won't be able
    # to convert X to float. 
    for i in range(0,len(data_numpy)): # number of rows
        horsepower = data_numpy[i][3]
        if horsepower == '?':
            rows_to_delete.append(i)
    
    for i in range(len(rows_to_delete)-1, -1, -1):
        row = rows_to_delete[i]
        data_numpy = np.delete(data_numpy, row, 0)
    
    X = data_numpy[:,1:8].astype("float")
    Y = data_numpy[:,0].astype("float")
    
    if use_dataset3:
        additional_col = data_numpy[:,9]
        
        # Execute one hot encoding on the new column. 
        encoder = LabelEncoder()
        additional_col_encoded = encoder.fit_transform(additional_col)
        additional_col_encoded = additional_col_encoded.reshape(-1, 1)
        # No need to create a dummy array as it's not the solution of a neural network. 
        
        X = np.append(X, additional_col_encoded, 1)#.astype("int")
        
elif use_dataset2:
    data_numpy = np.delete(data_numpy, 3, 1)
    
    X = data_numpy[:,1:7].astype("float")
    Y = data_numpy[:,0].astype("float")
```

```python
# Model definition

epochs = 200
batch_size = 10
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

data_shape_rows, data_shape_cols = X.shape

def baseline_model():
    model = Sequential()
    model.add(Dense(48, input_dim=data_shape_cols, activation="relu"))
    model.add(Dense(16, activation="relu"))
    model.add(Dense(1))
    
    model.compile(loss="mean_squared_error", optimizer="adam", metrics=["mse"])
    
    return model
```

```python
from tensorflow.keras.wrappers.scikit_learn import KerasRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import KFold, cross_val_score
from sklearn.pipeline import Pipeline

estimators = []
estimators.append(("standardization",StandardScaler()))
estimators.append(("mlp", KerasRegressor(build_fn = baseline_model, epochs=epochs, batch_size=batch_size)))

pipeline = Pipeline(estimators)

kfold = KFold(n_splits = 10, shuffle = True)
```

```python
# Train the model!

result = cross_val_score(pipeline, X, Y, cv=kfold, scoring='neg_mean_squared_error')
```

```python
print("Baseline: %.2f (%.2f) MSE" % (-result.mean(),result.std()))
```

```python
# For fun, execute the model (not on scaled data but whatever...) and look at the results. 

#model = baseline_model()

#model.fit(X, Y, epochs = epochs, batch_size=batch_size)

# Use pipline rather than just fitting a model without scaling.

model = pipeline.fit(X, Y)
```

```python
# Make predictions. 
predictions = model.predict(X)

# Create a new predictions dataframe
data_with_predictions = data.copy()

# In case we deleted rows, we need to delete them off our predictions dataframe. 
for i in range(len(rows_to_delete)-1, -1, -1):
    row = rows_to_delete[i]
    data_with_predictions = data_with_predictions.drop(row, 0)

# Append the predictions
data_with_predictions["mpg_predicted"] = predictions

# Do some rearranging of df columns to make it so the MPG is at the end. 
data_names_with_predictions = ["cylinders","displacement","horsepower","weight","acceleration","model_year","origin","car_name","mpg","mpg_predicted"]
data_with_predictions = data_with_predictions[data_names_with_predictions]

data_with_predictions
```

```python
# For fun, let's pop in my own truck's information and see
# what it spits out. 

# "Ford F-150 XLT 4.6L V8 4x4"
# Expected MPG: 14.0
additional_row_mpg = "14.0"

# Note: "acceleration" metric is not known. Utilized acceleration
# value of a ford f108 (mpg of 13.0) plus 1. 
additional_row_names = np.array(["cylinders","displacement","horsepower","weight","acceleration","model_year","origin"])
additional_row = np.array([[8,281,220,4478,11.0,70,1]])

additional_row_with_predictions = pd.DataFrame(additional_row, columns=additional_row_names)
additional_row_with_predictions["mpg"] = [additional_row_mpg]
```

```python
additional_prediction = model.predict(additional_row)

# Observation:
#
# Model has trained heavily on the model_year. It expects similar
# improvement in engine efficiency across years and thus places 
# heavy weight on the model_year attribute. For this particular
# experiment, setting the model year to 97 results in a predicted
# mpg of 42.242. Changing only the model year to 70 results in a
# mpg of 14.336. 
#
# There is also the rather obvious issue of years above the turn
# of the century. Using a model year of 100 for 2000 just makes 
# things confusing. The data should be updated to specify the 
# full year. 

additional_row_with_predictions["mpg_predicted"] = [additional_prediction]

additional_row_with_predictions
```
