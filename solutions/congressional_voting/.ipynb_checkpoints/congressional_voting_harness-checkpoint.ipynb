```python
# Congressional Voting Dataset

data_folder = "./data"
dataset_name = "house-votes-84.data"

import numpy as np
import pandas as pd
```

```python
# Import data

data_names = ["class","hi","wpcs","aotbr","pff","esa","rgis","astb","atnc","mm","i","scc","es","srts","c","dfe","eaasa"]
data_delimiter = ","

data = pd.read_csv(data_folder + "/" + dataset_name, names=data_names, delimiter=data_delimiter)
data
```

```python
# Data Analysis

import matplotlib.pyplot as plt

data_colors = {"republican":"red","democrat":"blue"}

def scatter_attribute(data, attribute, solution):
    fig, ax = plt.subplots()
    ax.title.set_text("congressional_voting dataset - " + attribute)
    graph_data = data.groupby(data[solution])[attribute]
    graph_data.hist(ax=ax) #alpha=0.4)
    #plt.scatter(data.index,data[attribute],color=data[solution].map(data_colors))
    #plt.show()

for col in data:
    scatter_attribute(data, col, "class")
```

```python
# Data Preprocessing

data_numpy = data.to_numpy()

data_rows, data_cols = data_numpy.shape

X = data_numpy[:,1:data_cols]
Y = data_numpy[:,0]

# Encode everything (y, n, ?, democrat/republican)
from sklearn.preprocessing import LabelEncoder

X_rows, X_cols = X.shape
encoder = LabelEncoder()

for i in range(0,X_cols):
    X[:,i] = encoder.fit_transform(X[:,i])
X = X.astype("float")
Y = encoder.fit_transform(Y).astype("float")
```

```python
# Model Spot Check - known algorithms

X_rows, X_cols = X.shape

nn_epochs = 70
nn_batch_size = 5
knn_n_neighbors=5
knn_metric="euclidean"
dt_criterion="entropy"
rf_criterion="gini"
rf_max_depth=None
rf_n_estimators=80
svm_C=0.1
svm_kernel="rbf"
gb_n_estimators=100
gb_learning_rate=0.1
gb_subsample=0.8
gb_criterion="friedman_mse"

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB, MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn_lvq import GlvqModel

def baseline_nn():
    model = Sequential()
    model.add(Dense(4, input_dim=X_cols, activation="relu"))
    model.add(Dense(1))
    model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
    return model
def baseline_knn():
    return KNeighborsClassifier(n_neighbors=knn_n_neighbors,metric=knn_metric)
def baseline_gnb():
    return GaussianNB()
def baseline_mnb():
    return MultinomialNB()
def baseline_dt():
    return DecisionTreeClassifier(criterion=dt_criterion)
def baseline_rf():
    return RandomForestClassifier(criterion=rf_criterion,n_estimators=rf_n_estimators,max_depth=rf_max_depth)
def baseline_gb():
    return GradientBoostingClassifier(criterion=gb_criterion,n_estimators=gb_n_estimators,learning_rate=gb_learning_rate,subsample=gb_subsample)
def baseline_svm():
    return SVC(C=svm_C,kernel=svm_kernel)
def baseline_lvq():
    return GlvqModel()

from sklearn.pipeline import Pipeline
from sklearn.model_selection import cross_val_score, KFold
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier

kfold = KFold(n_splits = 10, shuffle=True)

def model_pipleine(model, standardizer):
    estimators=[]
    estimators.append(("standardize",standardizer))
    estimators.append(("clf",model))
    return Pipeline(estimators)

def model_result(pipeline):
    return cross_val_score(pipeline, X, Y, cv=kfold, scoring="accuracy")

def model_print(result, model_name):
    print("Baseline " + model_name + ": %.2f (%.2f)" % (result.mean()*100,result.std()*100))

def model_pipeline_result(model, standardizer):
    pipeline = model_pipleine(model, standardizer)
    return model_result(pipeline)

def model_pipeline_result_print(model, standardizer, model_name):
    result = model_pipeline_result(model, standardizer)
    model_print(result, model_name)
    
```

```python
model_pipeline_result_print(baseline_knn(), StandardScaler(), "knn")
model_pipeline_result_print(baseline_gnb(), MinMaxScaler(feature_range=(0,1)), "gnb")
model_pipeline_result_print(baseline_mnb(), MinMaxScaler(feature_range=(0,1)), "mnb")
model_pipeline_result_print(baseline_dt(), StandardScaler(), "dt")
model_pipeline_result_print(baseline_rf(), StandardScaler(), "rf")
model_pipeline_result_print(baseline_svm(), StandardScaler(), "svm")
model_pipeline_result_print(baseline_lvq(), StandardScaler(), "lvq")
model_pipeline_result_print(baseline_gb(), StandardScaler(), "gb")
```

```python
result_nn = model_pipeline_result(KerasClassifier(build_fn=baseline_nn, epochs=nn_epochs, batch_size=nn_batch_size), StandardScaler())
                                  
```

```python
model_print(result_nn, "nn")
```

```python
# New algorithm - Linear Discriminant Analysis (LDA)

from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

def baseline_lda():
    return LinearDiscriminantAnalysis()

model_pipeline_result_print(baseline_lda(), StandardScaler(), "lda")
```
