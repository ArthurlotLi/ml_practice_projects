```python
# Harness wide constants

data_folder = "./data"
dataset_name = "housing.data"
```

```python
# Basic imports

import pandas as pd
import numpy as np
```

```python
# Read in data

data_names = ["crim", "zn", "indus", "chas", "nox", "rm", "age", "dis", "rad", "tax", "ptratio", "b", "lstat", "medv"]

data = pd.read_csv(data_folder + "/" + dataset_name, delim_whitespace=True, names=data_names)

data
```

```python
data["medv"].min(), data["medv"].max()
```

```python
#
# 2. Preparing the Data
# 

import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap, BoundaryNorm

bounds = [5, 10, 20, 30, 40, 50]
colors = ["red", "blue", "green", "yellow", "purple", "black"]
cmap = ListedColormap(colors)
norm = BoundaryNorm(bounds, len(colors))

def plot_column(dataset_name, col, data, solution_col):
    fig, ax = plt.subplots()
    ax.title.set_text(dataset_name + " dataset - " + col)
    sc = ax.scatter(data.index, data[col], c=data[solution_col], cmap=cmap, norm=norm)
    fig.colorbar(sc, spacing="proportional")
```

```python
for column in data:
    plot_column("housing", column, data, "medv")
```

```python
#
# Data Preprocessing
#

use_dataset_var_2 = False
use_dataset_var_3 = False

data_numpy = data.to_numpy()

X = data_numpy[:,0:13].astype("float")
Y = data_numpy[:,13].astype("float")
#Y = Y.reshape(-1, 1)

# Dataset variant (optional)
if use_dataset_var_2:
    X = np.delete(X, 3, 1)
    X = np.delete(X, 3, 1)
    X = np.delete(X, 6, 1)
    X = np.delete(X, 6, 1)
    X = np.delete(X, 6, 1)
elif use_dataset_var_3:
    X = np.delete(X, 11, 1) 
```

```python
#
# 3. Train Models
#

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

data_rows, data_cols = X.shape

def baseline_model():
    model = Sequential()
    model.add(Dense(26, input_dim=data_cols, activation="relu"))
    model.add(Dense(1))
    model.compile(loss="mean_squared_error", optimizer="adam",metrics=["mse"])
    return model
```

```python
from tensorflow.keras.wrappers.scikit_learn import KerasRegressor
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

epochs = 200
batch_size = 5

estimators = []
estimators.append(("standardize", StandardScaler()))
estimators.append(("mlp", KerasRegressor(build_fn = baseline_model, epochs=epochs, batch_size = batch_size)))

pipeline = Pipeline(estimators)

kfold = KFold(n_splits=10, shuffle=True)
```

```python
results = cross_val_score(pipeline, X, Y, cv=kfold)
```

```python
print("Baseline = %.2f (%.2f)" % (results.mean(), results.std()))
```

```python
# For fun, let's train a model and go a head and output the predictions. 

model = baseline_model()

model.fit(X, Y, epochs=epochs, batch_size=batch_size, verbose = 0)
```

```python
prediction = model.predict(X)
prediction

data_appended = data.copy()
data_appended['predicted_medv'] = prediction

# So we can peruse everything. 
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', -1)

data_appended
```
